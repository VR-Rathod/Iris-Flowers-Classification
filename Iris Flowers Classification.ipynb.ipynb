```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Select features (petal length and sepal length) and target
X = iris_df[['petal length (cm)', 'sepal length (cm)']]
y = iris_df['target']
```

```python
# Visualize the data
plt.scatter(X[y == 0]['petal length (cm)'], X[y == 0]['sepal length (cm)'], label='Iris-setosa')
plt.scatter(X[y == 1]['petal length (cm)'], X[y == 1]['sepal length (cm)'], label='Iris-versicolor')
plt.scatter(X[y == 2]['petal length (cm)'], X[y == 2]['sepal length (cm)'], label='Iris-virginica')

plt.xlabel('Petal Length (cm)')
plt.ylabel('Sepal Length (cm)')
plt.title('Iris Flowers Classification')
plt.legend()
plt.show()
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize and train the Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Predict using the trained model
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
print('Classification Report:\n', classification_report(y_test, y_pred))
```
